```python
import pandas as pd
test_d=pd.read_csv('kc_house_data.csv')
```

```python
test_d.head()
```

```python
# To check how many columns have missing values 
def show_missing():
    missing = test_d.columns[test_d.isnull().any()].tolist()
    return missing
```

```python
test_d[show_missing()].isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
le=LabelEncoder()
# Iterating over all the common columns in train and test
for col in test_d.columns.values:
       # Encoding only categorical variables
    if test_d[col].dtypes=='object':
       # Using whole data to form an exhaustive list of levels
        data=test_d[col].append(test_d[col])
        le.fit(data.values)
        test_d[col]=le.transform(test_d[col])
test_d.head()
```

```python
def normalize(test_d):
    res = test_d.copy()
    for feature in test_d.columns:
        
        if test_d[feature].dtype not in ['object', 'bool']:
            maxv = test_d[feature].max()
            minv = test_d[feature].min()
            res[feature] = (test_d[feature] - minv) / (maxv - minv)
    return res
test_d = normalize(test_d)
test_d.head()
```

```python
x = test_d[test_d.columns.drop(['price','id','date', 'lat', 'long'])]
y = test_d['price']
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size =0.2, random_state=0)
pd.DataFrame(X_test)
```

```python

from sklearn.ensemble import RandomForestRegressor
clf= RandomForestRegressor()
clf.fit(x, y)
```

```python
clf.score(x, y)
```

```python
from sklearn.linear_model import LinearRegression
clf1=LinearRegression()
clf1.fit(x,y)
```

```python
clf1.score(x,y)
```
